```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_theme(style="whitegrid")
```

```python

def plot(df):
    ax = sns.barplot(x="model_name", y="F1", data=df, orient="v", capsize=.2, palette="pastel")
    ax.set_xticklabels(ax.get_xticklabels(), 
                            rotation=60, 
                            horizontalalignment='right')
    ax.set(xlabel='')
```

```python
path_results = '../results/results-smote/'
```

## A_thrombosis_non_thrombosis_v5

```python
dset = 'A_thrombosis_non_thrombosis_v5'
test_size = 0.1
dataf = pd.read_csv(f'{path_results}{dset}_{test_size}.csv')
plot(dataf)
```

```python
dset = 'A_thrombosis_non_thrombosis_v5'
test_size = 0.15
dataf = pd.read_csv(f'{path_results}{dset}_{test_size}.csv')
plot(dataf)
```
