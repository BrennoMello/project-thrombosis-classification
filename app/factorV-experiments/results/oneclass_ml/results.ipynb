```python
import pandas as pd
```

```python
df = pd.read_csv('FV_final_dataset_clean_0.1.csv')
df
```

```python
for att in df['ATT'].unique():
    df_att = df[df.ATT == att]

    for model in df_att['model_name'].unique():
        df_att_model = df_att[df_att.model_name == model]
        print(f"att - {att} - model {model} - F1 {df_att_model.F1.mean()}")
```
